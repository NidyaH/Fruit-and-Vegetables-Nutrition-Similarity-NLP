```python
def get_pos(tag, wordnet):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN

def preprocess(text, wordnet):
  stemmer = SnowballStemmer('english')
  lemma = WordNetLemmatizer()

  text = word_tokenize(text)
  text = [w for w in text if w not in stopwords.words('english')]
  text = [w for w in text if w not in string.punctuation]
  text = [stemmer.stem(w) for w in text]

  tag = pos_tag(text)
  text = [lemma.lemmatize(w, get_pos(t, wordnet)) for (w,t) in tag]

  return ' '.join(text)
```

```python
def get_fruit(fruit: str):
    fruit = fruit.lower()
    matched = df[df['Name'].str.lower() == fruit]
    if not matched.empty:
        return matched['Benefit'].values[0]
    else:
        return None
```

```python
def vectorize(text1: str, text2: str):
    Benefit1 = get_fruit(text1)
    Benefit2 = get_fruit(text2)

    #check if get_fruit returned None (indicating not found)
    if Benefit1 is None or Benefit2 is None:
        print(f"One or both inputs not found in the dataset.")
        #print which one was not found
        if Benefit1 is None:
             print(f"'{text1}' not found.")
        if Benefit2 is None:
             print(f"'{text2}' not found.")
        return

    Benefit1_clean = preprocess(Benefit1, wordnet)
    Benefit2_clean = preprocess(Benefit2, wordnet)

    print("\n--- Preprocessed Benefits ---")
    print(f"{text1}: {Benefit1_clean}")
    print(f"{text2}: {Benefit2_clean}")

    #TF-IDF similarity
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([Benefit1_clean, Benefit2_clean])
    tfidf_score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

    #print both results
    print("\n--- Result ---")
    print(f"{text1.capitalize()} Benefit: {Benefit1}")
    print(f"{text2.capitalize()} Benefit: {Benefit2}")
    print(f"Lexical Similarity (TF-IDF): {tfidf_score:.2f}")

    semantic_similarity(Benefit1, Benefit2)
```

```python
def semantic_similarity(text1: str, text2: str):
    emb1 = model.encode(text1, convert_to_tensor=True)
    emb2 = model.encode(text2, convert_to_tensor=True)
    similarity = util.cos_sim(emb1, emb2).item()
    print(f"Semantic similarity: {similarity:.2f}")
```

```python
def show_top5_similar_each(fruit_name: str):
    benefit = get_fruit(fruit_name)
    if benefit is None:
        print(f"'{fruit_name}' tidak ditemukan dalam dataset.")
        return

    print(f"\nManfaat dari '{fruit_name}':\n{benefit}\n")
    all_benefits = df['Benefit'].tolist()
    all_embeddings = model.encode(all_benefits, convert_to_tensor=True)

    fruit_embedding = model.encode(benefit, convert_to_tensor=True)

    similarities = util.cos_sim(fruit_embedding, all_embeddings)[0].cpu().numpy()
    df['Similarity'] = similarities

    result = df[df['Name'].str.lower() != fruit_name.lower()]
    top5 = result.sort_values(by='Similarity', ascending=False).head(5)

    print(f"\nTop 5 buah/sayur mirip dengan '{fruit_name}':")
    print(top5[['Name', 'Benefit', 'Similarity']])
```

```python
def menu():
  while True:
    print("\n===Fruit/Vegetable Benefit Similarity===\n")
    print("1. Membandingkan kesamaaan antara 2 buah\n")
    print("2. Mencari 5 buah/sayur dengan kesamaan manfaat dengan buah yang dicari\n")
    print("3. Keluar\n")
    choice = input('>> ')

    if choice == '1':
      print('Membandingkan')
      text1 = input("Masukkan nama buah/sayur (1): ")
      text2 = input("Masukkan nama buah/sayur (2): ")
      vectorize(text1, text2)
    elif choice == '2':
      print("5 Buah/Sayur dengan Kesamaan Manfaat tertinggi")
      text3 = input("Masukkan nama buah/sayur: ")
      show_top5_similar_each(text3)
    elif choice == '3':
      return
    else:
      print("Invalid")

menu()
```
